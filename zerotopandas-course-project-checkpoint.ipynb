## Exploratory Analysis and Visualization

In this section we can make the data to be visualized with several barcharts and graphs for the further analysis of the udemy courses data

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Graph between number of subscribers and rating. It shows that how people select courses.

```python
data_df.plot(x='num_subscribers',y='rating')
```

```python
plt.scatter(data_df['num_subscribers'], data_df['rating'])
plt.show()
```

Here we can visualize the average rating of courses. 

```python
sns.distplot(data_df['avg_rating_recent'])
sns.distplot(data_df['avg_rating'])
```

```python
with sns.axes_style('white'):
    sns.jointplot(data_df['avg_rating_recent'], data_df['avg_rating'], data_df, kind='kde')
```

Graph between the price of course to the discount given on that course.

```python
plt.plot(data_df['price_detail__amount'],data_df['discount_price__amount'])
```

```python
sns.jointplot(data_df['avg_rating'], data_df['num_subscribers'], data=data_df, kind='reg');
```

Visual representation of the data if it has any null values in the datasets

```python
sns.heatmap(data_df.isnull())
```

## Asking and Answering Questions


#### Q1: How people select courses?

```python
plt.scatter(data_df['num_subscribers'], data_df['rating'])
plt.show()
```

So clearly people select courses based on their reviews. We can clearly see that courses with reviews 4+ have crossed marks of above 35k+ subs.

#### Q2: Have the courses lost it's valur over time ?

```python
sns.distplot(data_df['avg_rating_recent'])
sns.distplot(data_df['avg_rating'])
```

#### The answer to this is clearly no. As we can simply see more or less the courses maintain the reviews and few poor courses have even raised their value but it has not been declined.

#### Q3: Is there a profit made due to the discount given?

```python
plt.plot(data_df['price_detail__amount'],data_df['discount_price__amount'])
```

#### It has been seen the beta is roughly <= 1 which implies it has not been proved to make a very significant difference yet the sales has gone up as people find it cheap and hence easier to buy.

#### Q4: What are the top 10 courses based on rating?

```python
top_courses_df = data_df.sort_values('rating', ascending=False).head(10)
```

```python
top_courses_df
```

#### In the above cell we can see the top 10 courses based on their rating.

#### Q5: How many free courses are there?

```python
sns.countplot(x='is_paid', data=data_df);
```

```python
data_df['is_paid'].value_counts()
```

#### We can clearly see that only 496 courses are free in udemy as of the data. Most of the courses are paid.
